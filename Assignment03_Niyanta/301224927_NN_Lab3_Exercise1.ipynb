Using Data API - Loading and presprossing Data with tensorflow

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from scipy import stats
import tensorflow as tf
```

```python
column_names = ["First_Term_Gpa", "Second_Term_Gpa", "First_Language", "Funding", "School", "FastTrack", "Coop", "Residency", "Gender", "Previous_Education", "Age_Group", "High_School_Average_Mark", "Math_Score", "English_Grade", "FirstYearPersistence"]
niyanta_dataset = pd.read_csv("C:/Users/niyan/OneDrive/Desktop/Student_data.csv")
niyanta_dataset.columns = column_names
niyanta_dataset.head()
```

```python
# finding nan values each column
niyanta_dataset.isna().sum()
```

```python
# checking datatypes for each column
niyanta_dataset.dtypes
```

```python
# converting columns to numeric 
for column_name in ["First_Term_Gpa", "Second_Term_Gpa", "First_Language", "Previous_Education", "Age_Group", "High_School_Average_Mark", "Math_Score", "English_Grade"]:
    niyanta_dataset[column_name] = pd.to_numeric(niyanta_dataset[column_name], errors='coerce')
niyanta_dataset.head(10)
```

```python
niyanta_dataset.dtypes
```

```python
# filling the nan values with column mean

for col in niyanta_dataset.columns:
    niyanta_dataset[col] = niyanta_dataset[col].fillna(niyanta_dataset[col].mean())
print(niyanta_dataset.isnull().sum())
niyanta_dataset.head()
```

```python
# finding standard deviation 

std_scaler = StandardScaler()
std_scaler.fit(niyanta_dataset)
```

```python
#standardization using standard scaler

standard_data = std_scaler.transform(niyanta_dataset)
print(standard_data)
```

```python
#transformation

from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
pt.fit(niyanta_dataset)
pt.transform(niyanta_dataset)
```

```python
#load dataframe to tensors

tensor_data = tf.convert_to_tensor(niyanta_dataset)
print("tensor data", tensor_data)
```
